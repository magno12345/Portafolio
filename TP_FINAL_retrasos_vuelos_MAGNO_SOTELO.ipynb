##Modelo knn

> DAY_OF_MONTH y DAY_OF_WEEK pueden capturar patrones temporales en los retrasos de vuelos. Puede haber días de la semana o fechas específicas del mes que tienden a experimentar más retrasos.

> La distancia(DISTANCE) entre aeropuertos es relevante. Los vuelos más largos podrían estar más sujetos a retrasos debido a una variedad de factores, como escalas, condiciones meteorológicas, etc.

*  88,095 vuelos fueron correctamente clasificados como no retrasados (Verdaderos Negativos).

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix , accuracy_score, classification_report



x = flights_jan_2019[['DAY_OF_MONTH', 'DAY_OF_WEEK', 'DISTANCE','DEP_DEL15']]
y = flights_jan_2019['ARR_DEL15'].values.ravel()#targets
#creo una instancia de scaler
scaler = StandardScaler()
#escalo la x
x_escalada = scaler.fit_transform(x)
#conjuntos de entrenamiento y prueba
x_train, x_test, y_train, y_test = train_test_split(x_escalada, y, test_size=0.2, random_state=42)

# Escalo los datos

x_train_escalada = scaler.fit_transform(x_train)
x_test_escalada = scaler.transform(x_test)

def generate_clf_KNN(k, x_train, y_train):
    clasificadorKNN = KNeighborsClassifier(n_neighbors=k,metric="euclidean")
    clasificadorKNN.fit(x_train, y_train)
    return clasificadorKNN
#creo el modelo
knn = generate_clf_KNN(5,x_train_escalada,y_train)

from sklearn.model_selection import cross_val_score

# predictions son los target predecidos del modelo
predictions = knn.predict(x_test_escalada)
# Evaluación del modelo
accuracy = accuracy_score(y_test, predictions)
print(f'Precisión del modelo: {accuracy}')
# Reporte de clasificación
print('\nReporte de clasificación:')
print(classification_report(y_test, predictions))


#funcion para calcular la matriz de confusion
def matriz_confusion(y_test, predictions, nombre_model=""):
    # Calcula la matriz de confusión
    matrix = confusion_matrix(y_test, predictions)

    # Convierte la matriz de confusión a una lista de listas de cadenas
    confusion_matrix_str = [[str(value) for value in row] for row in matrix]

    # Crea un gráfico de matriz de confusión con Seaborn
    plt.figure(figsize=(4, 3))
    sns.heatmap(matrix, annot=True, fmt='g', linewidths=0.5, annot_kws={"size": 14},
                xticklabels=['No Retraso (0)', 'Retraso (1)'], yticklabels=['No Retraso (0)', 'Retraso (1)'])
    plt.xlabel('Predicciones del Modelo')
    plt.ylabel('Valores Verdaderos')
    plt.title(f'Matriz de Confusión de {nombre_model}')
    plt.show()


matriz_confusion(y_test, predictions, "KNN")
#print(confusion_matrix(y_test, predictions))
```

La **AUC-ROC** evalúa la capacidad de un modelo para distinguir entre las clases positivas y negativas. En un problema de clasificación binaria, como predecir si un vuelo estará o no retrasado, las clases son "positivo" (retrasado) y "negativo" (no retrasado). Cuanto mayor sea el AUC-ROC, mejor será la capacidad del modelo para discriminar entre estas clases.

```python
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.metrics import precision_score, recall_score, f1_score
#AUC-ROC
classifiers = [
    ('Random Forest', RandomForestClassifier(random_state=42)),
    ('Decision Tree', DecisionTreeClassifier(random_state=42)),
    ('K-Nearest Neighbors', KNeighborsClassifier()),
    ('Logistic Regression', LogisticRegression(random_state=42))
]

auc_roc_scores = []
for classifier_name, classifier in classifiers:
    classifier.fit(x_train_escalada, y_train)
    y_pred_proba = classifier.predict_proba(x_test)[:, 1]  # Probability of positive class
    auc_roc = roc_auc_score(y_test, y_pred_proba)
    auc_roc_scores.append((classifier_name, auc_roc))


auc_roc_scores = sorted(auc_roc_scores, key=lambda x: x[1], reverse=True)

print("AUC-ROC Scores:")
for classifier_name, auc_roc in auc_roc_scores:
    print(f"{classifier_name}: {auc_roc}")
```



*   **Random Forest** (Bosques Aleatorios): Un AUC-ROC de 0.8769 indica que el modelo Random Forest tiene una buena capacidad para discriminar entre vuelos con retraso y vuelos sin retraso. Cuanto más cercano a 1 sea el AUC-ROC, mejor es el rendimiento del modelo.
*   **K-Nearest Neighbors** (KNN): Con un AUC-ROC de 0.8665, el modelo KNN también presenta una capacidad sólida para la clasificación. Este valor sugiere que, en general, el modelo KNN puede distinguir eficazmente entre vuelos retrasados y no retrasados.
*  **Logistic Regression**  (Regresión Logística): Un AUC-ROC de 0.8537 indica que el modelo de Regresión Logística tiene un buen rendimiento, aunque ligeramente inferior a Random Forest y KNN. Aún así, es un valor sólido que sugiere que el modelo puede hacer buenas predicciones.
* **Decision Tree** (Árbol de Decisión): Con un AUC-ROC de 0.8472, el modelo de Árbol de Decisión también muestra una capacidad razonable para la clasificación. Aunque es ligeramente menor que otros modelos, sigue siendo un rendimiento aceptable.
